```python
from PIL import Image
img = Image.open(r'C:\Users\Rohit\Desktop\DIP-master\img.jpg')
pixels = img.load()
structuring_element = [
    [ 255 , 0,  255 ],
    [255, 0, 0],
    [ 0 , 0,  0],
]
print("Structurig element ")
print(structuring_element)
print("Original Image")
display(img)
rows = img.size[0]
cols = img.size[1]
def check(m1, m2):
  for i in range(len(m1)):
    for j in range(len(m1[0])):
      if (m1[i][j] != 0 )&  (m2[i][j] != 0):
        return True
  return False
  
for i in range(rows):
  for j in range(cols):
    if (i == 0) & (j == 0):
      pass
    elif i == rows-1 & j == cols-1:
      pass
    elif i == 0:
      pass
    elif j == 0:
      pass
    elif i == rows-1:
      pass
    elif j == cols-1:
      pass
    else:
      m1 = [
          [pixels[i-1, j-1][2], pixels[i-1, j][2], pixels[i-1, j+1][2]],
          [pixels[i, j-1][2], pixels[i, j][2], pixels[i, j+1][2]],
          [pixels[i+1, j-1][2], pixels[i+1, j][2], pixels[i+1, j+1][2]]
      ]
      if check(m1, structuring_element) == True:
        #true means hit 
        pixels[i, j] = (0, 0, 0)
print("After dilation")
display(img)
img = Image.open(r'C:\Users\Rohit\Desktop\DIP-master\img.jpg')
pixels = img.load()
```
