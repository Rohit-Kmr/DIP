```python
for i in range(img1.shape[0]):
  for j in range(img1.shape[1]):
    p2[i, j][0] = abs(p1[i, j][0] - p2[i, j][0])
    p2[i, j][1] = abs(p1[i,j][1]-p2[i,j][1])
    p2[i, j][2] = abs(p1[i, j][2] - p2[i, j][2])
imshow(p2)
```

```python
s = imread(r'C:\Users\Rohit\Desktop\DIP-master\img.jpg')
p=np.array(s)
print("To be sharpened")
imshow(s)
```
